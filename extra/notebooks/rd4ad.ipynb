# RD4AD Anomaly Detection: Reverse Distillation Pipeline

This notebook documents the workflow for the **Reverse Distillation for Anomaly Detection (RD4AD)** model implemented for the ADL 2025-2026 challenge. 

The RD4AD pipeline consists of:
1. **Reverse Distillation**: A frozen teacher (WideResNet-50) extracts features, which are reconstructed by a symmetric decoder through a one-class bottleneck.
2. **Multi-Scale Feature Comparison**: Anomalies are detected by measuring the cosine distance between teacher and decoder features at multiple scales.
3. **Multi-View Aggregation**: Combining signals from multiple views (5 per sample) to improve localization robustness.
4. **Dust Removal**: Post-processing to suppress small isolated noise blobs in the anomaly maps.
