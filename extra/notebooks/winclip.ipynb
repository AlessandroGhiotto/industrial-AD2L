# WinCLIP+ with Few-Shot Semantic Search

This notebook demonstrates the execution of the powerful **WinCLIP+** approach. It leverages:
1. **Zero-Shot Text Prompting**: Uses `openai/clip-vit-base-patch32` to find textual alignment between local patches and precise anomaly descriptions.
2. **Few-Shot Semantic Search**: Determines the specific `defect_type` of an unknown test sample by running K-NN on the global `CLS` token against the labeled anomaly examples.
3. **Memory Bank of the Normal**: Uses the `good` training images to build a FAISS memory bank to compute geometric distance to normality, which is fused with the text score.
